### Step 5: Outliers
####    Outlier detection:

```python
plt.figure(figsize=(20,10))
eda.boxplot()
```

#### Age, Fare and SibSp has outliers

```python
plt.scatter(eda['Fare'],eda['Age'])
```

#### Handling outliers of Age

```python
eda.boxplot(column='Age')
```

```python
eda['Age'].describe()
```

```python
IQR = eda['Age'].quantile(0.75)-eda['Age'].quantile(0.25) 
print(IQR)
```

```python
Upper_OutlierLimit = eda['Age'].quantile(0.75) + 1.5*IQR
Lower_OutlierLimit = eda['Age'].quantile(0.25) - 1.5*IQR 
print(Upper_OutlierLimit)
print(Lower_OutlierLimit) 
```

```python
OutlierValues = eda[(eda['Age']>=Upper_OutlierLimit) | (eda[ 'Age']<=Lower_OutlierLimit)]
```

```python
OutlierValues
```

```python
OutlierValues.reset_index(inplace=True)
```

```python
for i in (OutlierValues['index']):
    if eda['Age'][i]>=Upper_OutlierLimit:
        eda['Age'][i]=eda['Age'].quantile(0.95)
    else:
        eda['Age'][i]=eda['Age'].quantile(0.25)
```

```python
eda.boxplot(column='Age')
```

#### Handling outliers of Fare

```python
plt.figure(figsize=(12,7))
eda.boxplot(column='Fare')
```

```python
 eda['Fare'].describe() 
```

```python
IQR = eda['Fare'].quantile(0.75)-eda['Fare'].quantile(0.25) 
print(IQR)
```

```python
Upper_OutlierLimit = eda['Fare'].quantile(0.75) + 1.5*IQR
print(Upper_OutlierLimit)
```

```python
OutlierValues = eda[(eda['Fare']>=Upper_OutlierLimit)] 
```

```python
OutlierValues
```

```python
OutlierValues.reset_index(inplace=True)
```

```python
value=eda['Fare'].quantile(0.85)
value
```

#### Handling outliers for SipSb

# Various Classification Models and Accuracies

## Defining Variables X & y:

```python
cols = eda.columns
cols=[ 'Pclass','Age','SibSp', 'Parch','Fare','Sex_male', 'Embarked_Q', 'Embarked_S','Survived']
```

```python
eda = eda[cols]
eda
```

```python
X = eda.iloc[:,:-1].values
y = eda.iloc[:,-1].values
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

## Splitting the data in Training and Test set:

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
X_train.shape
```

## Building Logistic Regression Model:

```python
from sklearn.linear_model import LogisticRegression
# Representing LinearRegression as lr(Creating LinearRegression Object)
lr = LogisticRegression()

# fit the model to the training data
lr.fit(X_train, y_train)
```

```python
y_pred1 = lr.predict(X_test)
```

```python
y_test
```

```python
y_pred1
```

## Model Evaluation using Confusion Matrix:

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred1)
print(confusion)
```

```python
from sklearn.metrics import confusion_matrix
cmat=confusion_matrix(y_test,y_pred1)
print(cmat)
TP=cmat[1,1]
TN=cmat[0,0]
FP=cmat[0,1]
FN=cmat[1,0]
pd.DataFrame([TP,TN,FP,FN],index=['TP','TN','FP','FN'])
```

## Confusion Matrix Evaluation Metrics:

```python
from sklearn import metrics
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred1)*100,2),"%")
```

### Classification Error

```python
print (round(100-metrics.accuracy_score(y_test, y_pred1)*100,2),"%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```

### The accuracy obtained by building the model using Logistic Regression is 79.82%

## Find optimal value for k

```python
error = []
accuracy = []
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# Calculating error for K values between 
for i in range(1,100,2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))
    accuracy.append(accuracy_score(y_test, pred_i))
    print('k=',i,'accuracy=',accuracy_score(y_test, pred_i))
print(max(accuracy))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1,100,2), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1,100,2), accuracy, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Accuracy Score K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
```

#### The value of k for which we get highest accuracy is 23
